## Clase Pedido

#### Clase pedido con 3 atributos folio, producto, cantidad, 3 funciones leerdatos(), mostrardatos()
#### que sea interactivo ---> los input
#### Crear 3 objetos pedido1, pedido2
#### No usar while true, if, for, listas, diccionarios

![texto del vínculo](https://raw.githubusercontent.com/keylacigarroa128/ACT-26-Clase-Pedido-Keyla-0033/refs/heads/main/infoclasepedido.png)

```python
class Pedido:
    def __init__(self, folio = "", producto = "", cantidad = 0):
        self.folio = folio
        self.producto = producto
        self.cantidad = cantidad

    def leerdatos(self):
        self.folio = input("Ingrese el número de folio: ")
        self.producto = input("Ingrese el nombre del producto: ")
        self.cantidad = int(input("Ingrese la cantidad: "))

    def mostrardatos(self):
        print(f"Folio: {self.folio}")
        print(f"Producto: {self.producto}")
        print(f"Cantidad: {self.cantidad}")

# Creación de los 2 objetos
pedido1 = Pedido()
pedido2 = Pedido()

# Interacción secuencial
print("Datos del Pedido 1:")
pedido1.leerdatos()

print("\nDatos del Pedido 2:")
pedido2.leerdatos()

# Muestra de resultados
print("\n--- Información capturada ---")
pedido1.mostrardatos()
pedido2.mostrardatos()
print("software elaborado por Keyla Paola Cigarroa Palacios NL = 12")
```
